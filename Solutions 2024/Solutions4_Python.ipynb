# Problem 1, question f)

```python
T=1000
N=10
i=4
r=1.1
i_table=np.zeros((T,2))
for t in range(T):
    i_table[t,0]=t
    i_table[t,1]=i
    death=random()
    birth=random()
    if death<i/N and birth>r*i/(r*i+N-i): #An A dies and a B divides: i decreases by one
        i-=1
    elif death>i/N and birth<r*i/(r*i+N-i): #A B dies and an A divides: i increases by one
        i+=1
```

# Problem 1, question g)

```python
T=1000
N=10
i=4
r=1.1
i_table=np.zeros((T,2))
t=0 #now that we are using a while loop instead of a for loop, we need to explicitly initialize t
while i>0 and i<N:
    i_table[t,0]=t
    i_table[t,1]=i
    death=random()
    birth=random()
    if death<i/N and birth>r*i/(r*i+N-i): #An A dies and a B divides: i decreases by one
        i-=1
    elif death>i/N and birth<r*i/(r*i+N-i): #A B dies and an A divides: i increases by one
        i+=1
    t+=1 #now that we are using a while loop instead of a for loop, t does not increment automatically
i_table[t,0]=t #also save the last result
i_table[t,1]=i #also save the last result
i_table = i_table[~np.all(i_table == 0, axis=1)] #removes rows with just zeros
```

```python
plt.plot(i_table[:,0], i_table[:,1],'b.-')
ax = plt.gca()
ax.set_xlabel("Discrete time t")
ax.set_ylabel("Number i of A individuals")
plt.show()
```

# Problem 2, question c)

```python
r=1
M=1000
N_values=[10,12,15,20,25,50,100,200,500,1000]
probas_times_table=np.zeros((np.size(N_values),5))
for iN in range(np.size(N_values)):
    N=N_values[iN]
    probas_times_table[iN,0]=N
    results_table=np.zeros((M,2))
    for m in range(M):
        i=1 #we want to start with the same value of i for each replicate m; now i=1 initially
        t=0 #we need to initialize t for each replicate m
        while i>0 and i<N:
            death=random()
            birth=random()
            if death<i/N and birth>r*i/(r*i+N-i): #An A dies and a B divides: i decreases by one
                i-=1
            elif death>i/N and birth<r*i/(r*i+N-i): #A B dies and an A divides: i increases by one
                i+=1
            t+=1 #now that we are using a while loop instead of a for loop, t does not increment automatically
        if i==0:
            results_table[m,0]=0
        elif i==N:
            results_table[m,0]=1
        else:
            print('error...')
        results_table[m,1]=t
    probas_times_table[iN,1]=np.sum(results_table[:,0])/M #pA
    probas_times_table[iN,2]=np.mean(results_table[:,1]) #tf
    probas_times_table[iN,3]=np.mean(results_table[results_table[:,0]==1,1]) #tfA
    probas_times_table[iN,4]=np.mean(results_table[results_table[:,0]==0,1]) #tfB
```

```python
probas_times_table
```

```python
plt.plot(probas_times_table[:,0], probas_times_table[:,1],'bo-')
ax = plt.gca()
ax.set_xlabel("Population size N")
ax.set_ylabel("Fixation probability pA of type A")
plt.show()
```

```python
plt.plot(1/probas_times_table[:,0], probas_times_table[:,1],'ro',label='Data')
plt.plot(1/probas_times_table[:,0], 1/probas_times_table[:,0],'k--',label='y=x')
plt.legend()
ax = plt.gca()
ax.set_xlabel("1/N")
ax.set_ylabel("Fixation probability pA of type A")
plt.show()
```

```python
plt.plot(probas_times_table[:,0], probas_times_table[:,2],'bo-')
ax = plt.gca()
ax.set_xlabel("Population size N")
ax.set_ylabel("Mean fixation time tf of type A or type B")
plt.show()
```
