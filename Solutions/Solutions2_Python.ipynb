# Problem 2, question e)

```python
num_points = 1000
x_min, x_max = 0, 1
P_sick = np.linspace(x_min, x_max, num_points)
```

```python
alpha=0.787
beta=0.997
P_sick_if_pos = alpha*P_sick / (alpha*P_sick+(1-beta)*(1-P_sick))
```

```python
plt.plot(P_sick, P_sick_if_pos,'r.-',label='P(s|p)')
plt.legend()
ax = plt.gca()
ax.set_xlabel("P(s)")
ax.set_ylabel("P(s|result)")
plt.show()
```

```python
P_sick_if_neg = (1-alpha)*P_sick / (1-alpha*P_sick-(1-beta)*(1-P_sick))
```

```python
plt.plot(P_sick, P_sick_if_neg,'b.-',label='P(s|n)')
plt.legend()
ax = plt.gca()
ax.set_xlabel("P(s)")
ax.set_ylabel("P(s|result)")
plt.show()
```

```python
plt.plot(P_sick, P_sick_if_pos,'r.-',label='P(s|p)')
plt.plot(P_sick, P_sick_if_neg,'b.-',label='P(s|n)')
plt.plot(P_sick, P_sick,'k--',label='y=x')
plt.legend()
ax = plt.gca()
ax.set_xlabel("P(s)")
ax.set_ylabel("P(s|result)")
plt.show()
```
